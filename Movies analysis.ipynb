```python
#import libaries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')

from matplotlib.pyplot import figure

%matplotlib inline 
matplotlib.rcParams['figure.figsize']= (12,8) #Adjusts the size of plotted graphs.

# Read in data

df=pd.read_csv(r'C:\Users\MR LUCKY\Documents\Practice Data file\movies.csv')
```

```python
#Previewing read data

df.head()
```

```python
# Data exploration

# Checking if there are any empty cells

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
   
```

```python
# checking for Data types of columns

df.dtypes
```

```python
# Show max data
df.head()

```

```python
#creating Month column by extracting just the Month from released
df['Release-Month']= df['released'].astype(str).str[0:4]
df
```

```python
# Dropping a created Column

df.drop(['Year-Correct'],axis=1)
```

```python
#Ploting a scatter graph to show the correlation between Gross earnings and Budget 

plt.scatter(x=df['budget'], y=df['gross'])
plt.title("Budget vs gross")
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()
```

```python
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"Red"}, line_kws={"color":"Purple"})
```

```python
#Calculating correlation of the data set

df.corr(method='pearson')
```

```python
#Showing calculated Correlation in a heatmap graph
Correlation_Matrix=df.corr(method='pearson')
sns.heatmap(Correlation_Matrix, annot=True)
plt.title("Correlation of movies feature")
plt.xlabel('Movies features')
plt.ylabel('Movies features')
plt.show
```
